```python
TRAIN_DIR = '../dataset/task1/train'
VAL_DIR = '../dataset/task1/val'
TEST_DIR = '../dataset/task1/test'
```

```python
def transform(path):
    for subDir in os.listdir(path):
        for images in os.listdir(os.path.join(path,subDir)):
            if images.endswith('.jpg'):
                name = path + "/" + subDir + "/" + images
                image = cv2.imread(name)
                resized_image = cv2.resize(image, (256, 256))
                gray_img = cv2.cvtColor(resized_image, cv2.COLOR_RGB2GRAY)
                kernel = cv2.getStructuringElement(1, (17, 17))
                blackhat = cv2.morphologyEx(gray_img, cv2.MORPH_BLACKHAT, kernel)
                _, threshold = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)
                final_image = cv2.inpaint(resized_image, threshold, 1, cv2.INPAINT_TELEA)
                print(name)
                cv2.imwrite(name, final_image)
```

```python
#Train
transform(TRAIN_DIR)
```

```python
#Validation
transform(VAL_DIR)
```

```python
#Test
transform(TEST_DIR) 
```

```python
 def transform1(image):
     #image = cv2.imread(image)
     resized_image = cv2.resize(image, (256, 256))
     gray_img = cv2.cvtColor(resized_image, cv2.COLOR_RGB2GRAY)
     kernel = cv2.getStructuringElement(1, (17, 17))
     blackhat = cv2.morphologyEx(gray_img, cv2.MORPH_BLACKHAT, kernel)
     _, threshold = cv2.threshold(blackhat, 10, 255, cv2.THRESH_BINARY)
     final_image = cv2.inpaint(resized_image, threshold, 1, cv2.INPAINT_TELEA)
     return final_image
```
